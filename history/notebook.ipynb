```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
name = 'sfcn_pyramid_0_hist.csv'
hist = pd.read_csv(name)
hist[['loss','val_loss']].plot()
```

```python
fig, ax  = plt.subplots()

name = 'sfcn_pyramid_avg_0_hist.csv'
hist = pd.read_csv(name)
ax.plot(hist[['val_loss']], label='lr=1e-4')

name = 'sfcn_pyramid_avg_4_hist.csv'
hist = pd.read_csv(name)
ax.plot(hist[['val_loss']], label='lr=1e-6')

name = 'sfcn_pyramid_avg_plat_1_hist.csv'
hist = pd.read_csv(name)
ax.plot(hist[['val_loss']], label='ROP lr=1e-5')

ax.set_yscale('log')
ax.set_xlabel('epochs')
ax.set_ylabel('mse loss')
ax.grid()

fig.legend()
fig.suptitle('validation loss')

plt.show()
```
